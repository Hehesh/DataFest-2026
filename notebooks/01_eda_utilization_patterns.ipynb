```python
from pathlib import Path
import sys

project_root = Path.cwd()
if project_root.name == 'notebooks':
    project_root = project_root.parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.clean_encounters import clean_encounters, filter_diabetes_patients
from src.load_data import load_encounters
```

```python
encounters = clean_encounters(load_encounters(sample=True))
diabetes = filter_diabetes_patients(encounters)

(
    diabetes.groupby(['county_name', 'year'])[['is_ed', 'is_hospital_admittance', 'is_observation']]
    .mean()
    .round(3)
    .sort_index()
)
```
