# Build County Panel

## Purpose
Turn cleaned encounter-level records into a county-year panel that supports downstream scoring.

## Inputs
- `data/sample/synthetic_encounters.csv`
- `data/sample/synthetic_county_panel.csv`

## Outputs
- County-year utilization and context table

## Main Steps
1. Load data
2. Clean encounter records
3. Aggregate to county-year level
4. Attach county context
5. Brief interpretation

```python
from pathlib import Path
import sys

project_root = Path.cwd()
if project_root.name == 'notebooks':
    project_root = project_root.parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.build_county_panel import build_county_panel
from src.clean_encounters import clean_encounters
from src.load_data import load_county_panel, load_encounters
```

```python
encounters = clean_encounters(load_encounters(sample=True))
county_context = load_county_panel(sample=True)
county_panel = build_county_panel(encounters).merge(
    county_context[['county_fips', 'county_name', 'year', 'uninsurance_rate', 'uninsurance_rate_change']],
    on=['county_fips', 'county_name', 'year'],
    how='left',
)
county_panel.head()
```
